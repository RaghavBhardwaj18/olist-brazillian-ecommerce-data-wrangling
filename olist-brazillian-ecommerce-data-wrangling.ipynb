**Olist Brazilian e-commerce dataset analysis**

```python
# importing necessary libraries
import numpy as np
import pandas as pd

customers = pd.read_csv("olist_customers_dataset.csv")
orders = pd.read_csv("olist_orders_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")

order_items = pd.read_csv("olist_order_items_dataset.csv")
products_name_translated = pd.read_csv("product_category_name_translation.csv")

print(customers.head(), customers.shape, sep="\n")
```

```python
print(products.head(), products.shape, sep="\n")
```

```python
print(order_items.head(), order_items.shape, sep="\n")
```

```python
print(products_name_translated.head(), products_name_translated.shape, sep="\n")
```

Phase 1 - Merge Data

```python
# merging all tables and creating a master dataset
# process: (order + order_items) -> (+ products + products_translated) -> (+customers)

# merging orders dataset with order items dataset
orders_with_items = pd.merge(orders, order_items, on="order_id", how="inner")
print(orders_with_items.head(), orders_with_items.shape, sep="\n")
```

```python
# merging products and products_name_translated to have a resultant dataset with products information along with their translated category names
products_with_names_translated = pd.merge(products, products_name_translated, on="product_category_name", how="inner")
print(products_with_names_translated.head(), products_with_names_translated.shape, sep="\n")
```

```python
products_df = products_with_names_translated.loc[:,["product_id","product_category_name_english"]]
print(products_df.head(), products_df.shape, sep="\n")
```

```python
# merging products dataset with the just obtained orders resultant dataset through product id
orders_with_items_products = pd.merge(orders_with_items, products_df, on="product_id", how="inner")
print(orders_with_items_products.head(),  orders_with_items_products.shape, sep="\n")
```

```python
# merging customers data with the resultant dataset through customer id
master_df = pd.merge(orders_with_items_products, customers, on="customer_id", how="inner")
print(master_df.head(), master_df.shape, sep="\n")
```

```python
print(master_df.isna().sum())
```

Phase 2 - Feature Engineering

```python
orders_cust = master_df.loc[:,["order_id","customer_id","order_status","price","freight_value","product_category_name_english"]]
print(orders_cust.head(), orders_cust.shape, sep="\n")
```

```python
# calculating total value for every order as : total_value = price + freight_value
orders_cust["total_value"] = orders_cust["price"] + orders_cust["freight_value"]
print(orders_cust.head(), orders_cust.shape, sep="\n")
```

```python
# calculating total_value per order
orders_groupby_df = orders_cust.groupby("order_id")["total_value"].sum()
print(orders_groupby_df.head(), orders_groupby_df.shape, sep="\n")
```

```python
orders_groupby_df = orders_groupby_df.reset_index()
```

```python
print(orders_groupby_df.head(), orders_groupby_df.shape, sep="\n")
```

Phase 3 - Analysis

```python
print(orders_cust.head(), orders_cust.shape, sep="\n")
```

```python
# calculating total value per customer
total_per_cust = orders_cust.groupby("customer_id")["total_value"].sum().reset_index()
print(total_per_cust.head(), total_per_cust.shape, sep="\n")
```

```python
# calculating total revenue per category
total_per_category = orders_cust.groupby("product_category_name_english")["total_value"].sum().reset_index()
print(total_per_category.head(), total_per_category.shape, sep="\n")
```

```python
# identifying top customers
total_per_cust["rank_cust"] = total_per_cust["total_value"].rank(ascending=False, method="dense")
print(total_per_cust.head(), total_per_cust.shape, sep="\n")
```

```python
total_per_cust.sort_values(by="rank_cust", inplace=True)
print("top 5 customers:", total_per_cust.head(), total_per_cust.shape, sep="\n")
```

Phase 4 - Pivot Tables

```python
print(orders_cust.head())
```

```python
# category vs revenue
print(orders_cust.pivot_table(index="product_category_name_english", values="total_value", aggfunc="sum"))
```

```python
# city vs revenue
orders_cust_city = pd.merge(orders_cust, master_df[["order_id","customer_city"]], on="order_id", how="inner")
print(orders_cust_city.head(), orders_cust_city.shape, sep="\n")
```

```python
# getting city vs revenue pivot table
print(orders_cust_city.pivot_table(index="customer_city", values="total_value", aggfunc="sum"))
```

```python
# customer vs category
print(orders_cust_city.pivot_table(index="customer_id", values="product_category_name_english", aggfunc="sum"))  # aggfunc="sum" can result to some issues here since it's being used on non-numeric (textual) data
```

Phase 5 - Reshaping

```python
# converting pivot -> long format
# creating pivot of total value by category and city
prod_value_pivot_df = orders_cust_city.pivot_table(index="product_category_name_english", columns="customer_city", values="total_value", aggfunc="sum", fill_value=0)
print(prod_value_pivot_df)
```

```python
# converting it to long format
prod_value_pivot_df = orders_cust_city.pivot_table(index="product_category_name_english", columns="customer_city", values="total_value", aggfunc="sum")
prod_value_melt_df = prod_value_pivot_df.reset_index().melt(id_vars=["product_category_name_english"], var_name="city", value_name="total_value").dropna(subset=["total_value"])
print(prod_value_melt_df)
```
